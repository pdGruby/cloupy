```python
import cloudy as cl
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('sudety_2.csv')
df = df.drop('Unnamed: 0', axis=1)
df = df.set_index('Stacja')
```

```python
station = 'ŚNIEŻKA'

data_for_wl = df[df.index == station]
data_for_wl = data_for_wl.loc[:, ['Miesiąc', 'Średnia temperatura miesięczna [°C]', 
                                  'Miesieczna suma opadów [mm]', 
                                  'Absolutna temperatura maksymalna [°C]', 
                                  'Absolutna temperatura minimalna [°C]']]
#wld(data_for_wl, years_range=range(1984, 2015), station_name='Śnieżka', language='POL', lat_lon=(50.74, 15.74), above_sea_level=1609)
```

```python
#cl.choose_diagStyle('retro')
cl.choose_diagStyle('default')
cl.WalterLieth(data_for_wl, years_range=range(1984, 2015), 
               station_name='Śnieżka', 
               lat_lon=(50.74, 15.74), 
               above_sea_level=1609).draw(language=None)
#plt.savefig('def.jpg')
```
